```python
from pathlib import Path

import polars as pl

project_path = Path.cwd() / ".." / "output" / "project"
```

```python
ricu_analysis = pl.read_json("output.json")
ricu_analysis
```

```python
faulty = []
```

```python
for e in ricu_analysis:
    e = e[0]
    if e["is_equal_in_ricu"]:
        cnpt_directory = project_path / "workspace" / "concept" / e["name"] / "1.0.0"
        # print(cnpt_directory)
        cnpt_paths = [str(p) for p in cnpt_directory.iterdir() if p.is_file()]
        cnpt =  pl.scan_parquet(cnpt_paths[0])
        cnpt = cnpt.collect()
        # print(e["in_ricu_df_size"] == len(cnpt), e["name"], ":", e["in_ricu_df_size"], len(cnpt))
        if not e["in_ricu_df_size"] == len(cnpt):
            faulty.append(e)
            print(e["name"])
        # print()
```

```python
for e in ricu_analysis:
    e = e[0]
    if not e["is_equal_in_ricu"]:
        print(e["name"], ":", e["in_ricu_df_size"])
        faulty.append(e)
```
