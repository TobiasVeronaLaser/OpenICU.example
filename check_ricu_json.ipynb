```python
def raise_dict(target, key): 
    if not target.get(key):
        target[key] = 0
    target[key] += 1
```

```python
print(f"key_count: {len(data.keys())}")
print(f"{data.keys()}")
```

```python
diff_keys = {}
for e in data.keys():
    for key in data[e].keys():
        if not diff_keys.get(key):
            diff_keys[key] = 0
        diff_keys[key] += 1

print(diff_keys)
```

```python
sorted_diff_keys = []
for key, value in diff_keys.items():
    sorted_diff_keys.append((key, value))
sorted_diff_keys = sorted(sorted_diff_keys, key=lambda x: x[1], reverse=True)
```
