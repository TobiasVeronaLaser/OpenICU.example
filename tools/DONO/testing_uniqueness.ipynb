```python
concepts_miiv = []
concepts_miiv_not_used = []
concepts_remaining = []
cpt_no_src_nor_miiv = []
```

```python
print("miiv\n")
for col in df.columns:
    if (concept := df[col][0]).get("sources") is None or "miiv" not in concept["sources"]:
        cpt_no_src_nor_miiv.append(concept)
        continue
    concept = df[col][0]
    print("category:\t", concept["category"], "\nshortcut:\t", col, "\ndescription:\t", df[col][0]["description"])
    miiv = concept["sources"]["miiv"]
    for e in miiv:
        if e.get("table") is None:
            print("No table")
            continue
        print("\ttable:\t", e["table"], "\n\tsub_var:\t", e.get("sub_var"), "\n\tval var:\t", e.get("val var"))
        print("\t\treg:\t", e.get("regex") is not None, "\n\t\tid:\t", e.get("ids") is not None)
    print("\n")
    #print(col, "\n", df[col][0]["sources"]["miiv"][0].keys())
    
```

```python
ricu_cpt_names = [(cpt_abbr, df[cpt_abbr][0]["description"])for cpt_abbr in df.columns]
ricu_cpt_names
```

```python
ricu_clses: set = set({})
for cpt_abbr in df.columns:
    ricu_cls = df[cpt_abbr][0].get("class")
    ricu_clses.add(ricu_cls if not isinstance(ricu_cls, list) else tuple(ricu_cls))
ricu_clses
```

```python
ricu_srcs = set({})
for cpt_abbr in df.columns:
    temp_ricu_srcs = df[cpt_abbr][0].get("sources")
    if not temp_ricu_srcs:
        ricu_srcs.add(None)
        continue
    ricu_srcs = ricu_srcs.union(set(list(temp_ricu_srcs.keys())))
ricu_srcs
```

```python
ricu_ctgs = set({})
for cpt_abbr in df.columns:
    reicu_ctg = df[cpt_abbr][0].get("category")
    ricu_ctgs.add(reicu_ctg)
ricu_ctgs
```
